```python
import matplotlib.pyplot as plt
from pandas import Series
from pandas.tools.plotting import lag_plot
%matplotlib inline
```

************************************************************************************************************************************************ SELECIONE TRÊS SÉRIES TEMPORAIS ***************************************************** *******************************************************************************************************************

```python
dados  = Series.from_csv('CSVs/international-airline-passengers.csv', sep=';', header=0)
dados1 = Series.from_csv('CSVs/monthly-traffic-fatalities-in-on.csv', sep=';', header=0)
dados2 = Series.from_csv('CSVs/passenger-miles-mil-flown-domest.csv', sep=';', header=0)
```

*****************************************************************************************************************************************************  Plotando os gráficos  ********************************************************* ****************   Linha                                                                         ****************** ****************   Pontos                                                                        ******************

```python
dados.plot()
dados1.plot()
dados2.plot()
```

```python
dados.plot(style='k.')
dados1.plot(style='r.')
dados2.plot(style='b.')
```

```python
dados.hist()
dados1.hist()
dados2.hist()
```
