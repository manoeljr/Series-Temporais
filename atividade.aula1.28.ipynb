```python
import matplotlib.pyplot as plt
from pandas import Series
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
%matplotlib inline
```

************************************************************************************************************************************************ SELECIONE TRÊS SÉRIES TEMPORAIS ***************************************************** *******************************************************************************************************************

```python
dados  = Series.from_csv('CSVs/international-airline-passengers.csv', sep=';', header=0)
dados1 = Series.from_csv('CSVs/monthly-traffic-fatalities-in-on.csv', sep=';', header=0)
dados2 = Series.from_csv('CSVs/passenger-miles-mil-flown-domest.csv', sep=';', header=0)
```

************************************************************************************************************************************************      PLOTANDO AS SERIES         ***************************************************** *******************************************************************************************************************

```python
dados.plot()
```

```python
dados1.plot()
```

```python
dados2.plot()
```

```python
decomposicao_dados  = seasonal_decompose(dados, model="additive")
decomposicao_dados1 = seasonal_decompose(dados1, model="additive")
decomposicao_dados2 = seasonal_decompose(dados2, model="additive")
```

```python
decomposicao_dados.plot()
decomposicao_dados1.plot()
decomposicao_dados2.plot()
```
